```python
df = pd.read_csv('french_dictionary.csv', header=None)
df.columns=['english', 'french']
df.index=df.english
df
```

```python
find_words = open('find_words.txt')
from collections import defaultdict 
word_dict = defaultdict(int)
for line in find_words.readlines():
  line_ = line[:-1]
  if line_ in df.index:
    word_dict[line_]+=1
  else:
    print(line_)
```

```python
sha_file = open('t8.shakespeare.txt').readlines()
out_file = open('output_file.txt', 'w')
```

```python
from collections import defaultdict 
txt_word_dict = defaultdict(int)
```

```python
%%time 
#for measuring time taken to run the code
for line in sha_file:
  for word in line.split():
    if word in df.index:
      txt_word_dict[word]+=1 #count words 
      line = line.replace(word, df.loc[word, 'french']) #get french word and replace it in the paragraph
  out_file.writelines(line) # creating a new file
```

```python
count_df = pd.DataFrame.from_dict(dict(txt_word_dict),  orient='index', columns=['count'])
count_df.to_csv('word_count.csv')
```
